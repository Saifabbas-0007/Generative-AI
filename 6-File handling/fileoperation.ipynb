File Operation: Read & Write Files
A file handling is a crucial part of any programming language.
Python provides built-in function and methods to read from and write to files,both text and binary

```python
## read a whole files
with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
# read a file line by line

with open('example.txt','r') as file:
    for line in file:
     print(line)
```

```python
with open('example.txt','r') as file:
    for line in file:
     print(line.strip())
```

```python
 # write a file that should overwrite the sentence
 
with open('example.txt','w')as file:
    file.write("hello world!\n")
    file.write("this is a new line. ")
    
```

```python
# write a file without overwrite

with open('example.txt','a') as file:
    file.write("Append function is used here!\n")
```

```python
# writing a list of lines to a file:
list=[ 'First line \n ','Second line \n ','Third line\n']
with open('example.txt','a') as file:
    file.writelines(list)
```

```python
#binary files
#writing to binary files

data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb')as file:
    file.write(data)
```

```python
# reading a binary files

with open('example.bin','rb')as file:
    content= file.read()
    print(content)
```

```python
## read the content from source txt file and write to a destination txt file

with open('example.txt','r') as source_file:
    content= source_file.read()
with open('destination.txt','w') as destination_file:
    destination_file.write(content)    
```

```python
## writing and the reading the file:

with open('example.txt','w+')as file:
    file.write("hello world\n")
    file.write("this is a new line \n")
    
    file.seek(0)
    content= file.read()
    print(content)
```
