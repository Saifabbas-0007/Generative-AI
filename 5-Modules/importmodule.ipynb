In Python , Modules and Packages helps organize and reuse the code

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1, 2, 3, 4, 5])
```

```python
from math import *
print(sqrt(36))
print(pi)
```

```python
from package.math import addition
result = addition(5, 10)
print(result)
```

```python
from package import math
print(math.addition(5, 10))
```
