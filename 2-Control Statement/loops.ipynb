```python
## while loops iterate until a condition is true:

count = 0
while count < 5:
    print(count)
    count += 1
print("done")    
```

```python
## loop control statements
## break statement is used to exit a loop prematurely when a certain condition is met. It allows you to terminate the loop and continue with the rest of the program.

str = "Python"
for i in str:
    if i == "h":
        break
    print(i)
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continue statement is used to skip the current iteration of a loop and move on to the next iteration. When the continue statement is encountered, the loop will skip the remaining code inside the loop for the current iteration and proceed to the next iteration.

for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
str = "MacBook Air"
for i in str:
    if i == " ":
        continue
    print(i)
```

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
 ## pass statement is used as a placeholder when you want to define a block of code but don't want to execute any code in that block. It allows you to create an empty block of code that can be filled in later.
 
 for i in range (5):
     if i == 3:
         print("the number is :",i)
         pass
     print(i)
```

```python
## nested loops
for i in range(3):
    for j in range(2):
        print("i:", i, "j:", j)  #or print(f"i: {i} j: {j}")
```

```python
# example :cakculate the sum of the first n natural numbers using while and for loop

n = 10 
sum = 0
count = 1

while count <= n:
    sum = sum + count
    count = count + 1
    print("the sum of first 10 Natural numbers is:", sum)
```

```python
n = 10
sum = 0
for i in range (1, n+1):
    sum += i
print("the sum of first 10 Natural numbers is:", sum)
```

```python
 ## prime number between 1 to 100
 
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
                
```
