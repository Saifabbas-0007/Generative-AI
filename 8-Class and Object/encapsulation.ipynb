### Encapsulation :


Encapsulation involves bundling data and methods that operates on data within a single unit. 


Encapsulation is the concept of wrapping the data (variables) and methods (function) together a single unit

```python
### Encapsultion with getter and setter
### public ,private and protected variable

class Person:
    def __init__(self,name,age):
        self.name = name    ##public variable
        self.age = age      ## public variable
        
person = Person("saif",23)
print(person.name)
```

```python
class Person:
    def __init__(self,name,age,gender):
        self.__name = name    ##private variable
        self.__age = age      ##private variable
        self.gender = gender
```

```python
class Person:
    def __init__(self,name,age,gender):
        self._name = name    ##protected variable
        self._age = age      ##protected variable
        self.gender = gender
        
class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)
        
employee = Employee("saif",23,"Male")
print(employee._name)
print(employee._age)
print(employee._name,employee._age,)
print(employee.gender)
```

```python
## Encapsulation with getter and setter method 

class Person:
    def __init__(self,name,age):
        self.__name = name #private variable
        self.__age = age  # private variable
        
    ## getter method for name
    def get_name(self):
        return self.__name
    
    ## setter method for name
    def set_name(self,name):
        self.__name = name
        
    ## getter method for age
    def get_age(self):
        return self.__age
    
    ## setter method for name
    def set_age(self,age):
        if age >0:
            self.__age = age
        else:
            print("Age cannot be negative")
            
person = Person("saif",23)

## Accessing and Modifying private variable using setter and getter

print(person.get_name())
print(person.get_age())

person.set_age(24)

print(person.get_age())

person.set_age(-23)
```
