### Magic Methods

Magic method in Python is also called as Dunder Method(duble underscore method) are special method that start and ends with double underscore. These methods enable you to define the behaviour of object for built-in operation such arithmetic operation ,comparison and more...

### Magic Methods 

Magic method are predefined methods in python that you can override to change the behaviour of your object.


 


```python
'''
__init__: Initialize a new instance class.
__str__: Returns a string representation of an object.
__repr__: Returns a official string representation of an object.
__len__: Returns a length of an object.
__getitem__: Gets an item from a container.
__setitem__: sets an item from a container.



'''
```

```python
## Basic method 

class Person:
    def __init__(self,name, age):
        self.name = name
        self.age = age
        
person = Person("saif",23)
print(person)
```

```python

class Person:
    def __init__(self,name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return(f" {self.name} is {self.age} years old... ")
        
person = Person("saif",23)
print(person)
```

```python
class Person:
    def __init__(self,name, age):
        self.name = name
        self.age = age
    def __repr__(self):
        return (f" name = {self.name} age = {self.age}")
        
person = Person("saif",23)
print(repr(person))
```

### operator overloading.

operator overloading allow you to define the behavior of operator(+,-,*,etc) for custom objects.

```python
'''
__add__
__sub__
__mul__
__truediv__
__eq__
__lt__
__gt__

'''
```

```python
class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def __add__(self,other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self,other):
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self,other):
        return Vector(self.x * other.x, self.y * other.y)
    
    def __eq__(self,other):
        return Vector(self.x == other.x, self.y == other.y)
    
    def __repr__(self):
        return f" Vector ({self.x},{self.y})"
    
v1 = Vector(2,3)
v2 = Vector(4,5)

print(v1+v2)
print(v1-v2)
print(v1*v2)
print(v1==v2)

```
