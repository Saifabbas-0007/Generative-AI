Exception handling: In python helps you to handle error gracefully and take corrective action without stoping the execution of the program . try,except,else,finally


What are Exception:


Exception are events that disrupt the flow of program, they occur when an error is encounter during program execution

```python
try:
    a = b
except:
    print('the variable is not assinged')
```

```python
try:
    num = int(input("enter a number"))
    result = 10/num
except ValueError :
    print("This is not a valid number")
except ZeroDivisionError:
    print("Provide the number greater than zero")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
```

```python
try:
    num = int(input("enter a number"))
    result = 10/num
except ValueError :
    print("This is not a valid number")
except ZeroDivisionError:
    print("Provide the number greater than zero")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
finally:
    print("Execution complete!")
```
