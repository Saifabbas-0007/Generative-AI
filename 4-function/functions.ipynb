```python
## why functions?
num = 24
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
    
```

```python
def check_even_odd(num):
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
## calling a function
check_even_odd(24)
check_even_odd(25)
```

```python
### variable length arguments
### positional and keyword arguments

# positional arguments
def numbers(*args):
    for num in args:
        print(num)
```

```python
numbers(1,2,3,4,5)
```

```python
## keyword arguments : all the parameters are passed as key=value pairs

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(name="Alice", age=30, city="New York")
print_details(name="saif", age=25, city="Cairo")
```

```python
## combination of positional and keyword arguments
def mixed_args(*args, **kwargs):
    print("Positional args:", args)
    print("Keyword args:", kwargs)
mixed_args(1, 2, 3, 4, name="Alice", age=30)
```

```python
def mixed_args(*args, **kwargs):
    for num in args:
        print(f"Positional arg: {num}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```
