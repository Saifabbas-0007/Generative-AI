```python
a = 10
b = 5

add_result = a+b
sub_result = a-b
mul_result = a*b
flo_result = a/b
div_result = a//b
mod_result = a%b
exponent_result = a**b
print("Addition:", add_result)
print("Subtraction:", sub_result)
print("Multiplication:", mul_result)
print("Floor Division:", flo_result)
print("Division:", div_result)
print("Modulo:", mod_result)
print("Exponentiation:", exponent_result)
```
